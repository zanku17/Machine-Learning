```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
pd.set_option("display.max_columns",None)
```

```python
data = sklearn.datasets.load_breast_cancer()
```

```python
data
```

```python
# Loading the data to a pandas data frame

df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
# Adding the target column to the data frame

df["label"] = data.target
df.tail()
```

```python
# Displaying the number of rows and column

df.shape
```

```python
df.info()
```

```python
# Checking the missing values

df.isnull().sum()
```

```python
# Statistical measures of Data

df.describe()
```

```python
# Checking the distribution of Target Variable
# 1 - Benign, 2 - Maliganant

df.label.value_counts()
```

```python
df.groupby("label").mean()
```

```python
# Spliting the data into Features and Target

X = df.drop("label", axis=1)
y = df.label
```

```python
X.head()
```

```python
y.head()
```

```python
# Spliting the data into Train and Test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = LogisticRegression()
```

```python
# Training the model using training Data

model.fit(X_train, y_train)
```

```python
# Finding the accuracy on training data

y_train_pred = model.predict(X_train)
```

```python
y_train_pred
```

```python
training_data_accuracy = accuracy_score(y_train, y_train_pred)
print("Accuracy on Training Data = ", training_data_accuracy)
```

```python
# Finding the accuracy on test data

y_test_pred = model.predict(X_test)
y_test_pred
```

```python
test_data_accuracy = accuracy_score(y_test, y_test_pred)
print("Accuracy on Test Data = ", test_data_accuracy)
```
